```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# read_csv
df = pd.read_csv('/content/car_details_from_car_dekho - car_details_from_car_dekho.csv')
df
```

```python
# Create a new column 'year_old' and drop the 'year' column
df['year_old'] = 2023 - df['year']
df = df.drop(columns=['year'])
df
```

```python

df = df.loc[:, df.nunique() > 1]
df
```

```python
#target column
target_col = 'selling_price'
X = df.drop(columns=[target_col])
y = df[target_col]
```

```python
#test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
print(X_train.columns)
```

```python
#Encoding
categorical_cols = ['name','fuel','seller_type', 'owner']
encoder = OrdinalEncoder()
X_train[categorical_cols] = encoder.fit_transform(X_train[categorical_cols])
X_test[categorical_cols] = encoder.transform(X_test[categorical_cols])
```

```python
# Random Forest
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
# Finding MAE
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error:', mae)
```
